**1.Score-Linear regression**

```python
# Import necessary libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Load your 'score' dataset (replace 'score_dataset.csv' with your actual dataset)
score_data = pd.read_csv('score.csv')
score_data.head()
# Identify independent variables (X) and dependent variable (y)
# Replace 'target_column' with the name of your dependent variable
X = score_data.drop('Scores', axis=1)
y = score_data['Scores']

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=23)

# Create a Linear Regression model
linear_model = LinearRegression()

# Fit the model to the training data
linear_model.fit(X_train, y_train)


# Predictions
y_pred = linear_model.predict(X_test)

#Actual and predicted
y_test = y_test.ravel()
y_pred = y_pred.ravel()
actual_pred = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
print(actual_pred)

# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("Training score: ",linear_model.score(X_train,y_train))
print("Testing score: ",linear_model.score(X_test,y_test))

print('Accuracy:',linear_model.score(X,y)*100)


# Print results
print("Mean Squared Error:", mse)
print("R-squared Score:", r2)

#Graph
import matplotlib.pyplot as plt
sap=pd.DataFrame({'Actual': y_test, 'Predicted': y_pred}).head(10)
sap.plot(kind='bar')
plt.grid(which='major',linewidth='0.8')
plt.grid(which='minor',linewidth='0.8')
plt.show()
```

**6.Boston-SVR**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error, r2_score

# Load data
sj = pd.read_csv('Boston.csv') # Replace 'your_dataset.csv' with your actual dataset

# Remove outliers
columns_to_process = ['crim', 'zn', 'rm', 'black', 'lstat', 'medv']
for col in columns_to_process:
    q1 = sj[col].quantile(0.25)
    q3 = sj[col].quantile(0.75)
    iqr = q3 - q1
    ul = q3 + 1.5 * iqr
    ll = q1 - 1.5 * iqr
    sj = sj[(sj[col] < ul) & (sj[col] > ll)]

# Convert selected columns to int
columns_to_convert = ['crim', 'zn', 'indus', 'nox', 'rm', 'age', 'dis', 'ptratio', 'black', 'lstat', 'medv']
sj[columns_to_convert] = sj[columns_to_convert].astype(int)

# Prepare data for modeling
x = sj.iloc[:, 1:14].values
y = sj.iloc[:, 14].values.reshape(-1, 1)

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)

# Train SVR model
reg = SVR()
reg.fit(x_train, y_train)

# Make predictions
y_pred = reg.predict(x_test)

# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print('Mean Squared Error:', mse)
print('R-squared Score:', r2)

# Visualize predictions
actual_pred = pd.DataFrame({'Actual': y_test.ravel(), 'Predict': y_pred.ravel()}).head(10)
actual_pred.plot(kind='bar')
plt.grid(which='major', linewidth='0.8')
plt.grid(which='minor', linewidth='0.8')
plt.xticks(rotation=90)
plt.show()

# Model scores
print("Training Score:", reg.score(x_train, y_train))
print("Testing Score:", reg.score(x_test, y_test))
print("Accuracy:", reg.score(x, y) * 100)
```

**8.Iris-DT,RF,GNB**

```python
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, mean_squared_error, r2_score
import seaborn as sns
import matplotlib.pyplot as plt

# Load the iris dataset
iris = load_iris()
X, y = iris.data, iris.target

# Split the train and test dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=23)

# Define classifiers
classifiers = [
    ("Decision Tree", DecisionTreeClassifier(random_state=0)),
    ("Random Forest", RandomForestClassifier(random_state=0)),
    ("Gaussian Naive Bayes", GaussianNB())
]

for name, clf in classifiers:
    # Train the classifier
    clf.fit(X_train, y_train)

    # Make predictions
    y_pred = clf.predict(X_test)

    # Calculate accuracy
    acc = accuracy_score(y_test, y_pred)
    print(f"{name} model accuracy (in %): {acc * 100}")

    # Display confusion matrix and classification report
    conf_matrix = confusion_matrix(y_test, y_pred)
    class_report = classification_report(y_test, y_pred)
    print(f"\n{name} Classifier:")
    print("Confusion Matrix:")
    print(conf_matrix)
    print("Classification Report:")
    print(class_report)

    # Visualize Confusion Matrix
    plt.figure(figsize=(8, 6))
    sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=iris.target_names, yticklabels=iris.target_names)
    plt.title(f'Confusion Matrix - {name} Classifier')
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.show()

    # Actual and Predicted values
    y_test = y_test.ravel()
    y_pred = y_pred.ravel()
    actual_pred = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
    print(actual_pred)

    # Evaluate the model
    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)

    print(f"\n{name} Model Evaluation:")
    print("Mean Squared Error:", mse)
    print("R-squared Score:", r2)

    # Print results
    print("Mean Squared Error:", mse)
    print("R-squared Score:", r2)

    # Bar graph for the first 10 samples
    sap = pd.DataFrame({'Actual': y_test[:10], 'Predicted': y_pred[:10]})
    sap.plot(kind='bar')
    plt.grid(which='major', linewidth='0.8')
    plt.grid(which='minor', linewidth='0.8')
    plt.xticks(rotation=90)
    plt.title(f'Actual vs Predicted - {name} Classifier')
    plt.show()
```

**12.RFC-Breastcancer**

```python
import pandas as pd
from sklearn.datasets import load_breast_cancer
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix, mean_squared_error, r2_score
import seaborn as sns
import matplotlib.pyplot as plt

# load the breast cancer dataset
X, y = load_breast_cancer(return_X_y=True)

# split the train and test dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=23)

# Random Forest Classifier
rf_clf = RandomForestClassifier(random_state=0)
rf_clf.fit(X_train, y_train)

# Prediction
y_pred_rf = rf_clf.predict(X_test)

# Actual and predicted
y_test = y_test.ravel()
y_pred = y_pred_rf.ravel()
actual_pred = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred_rf})
print(actual_pred)

# Evaluate the model
mse = mean_squared_error(y_test, y_pred_rf)
r2 = r2_score(y_test, y_pred_rf)

print("Training score: ",rf_clf.score(X_train, y_train))
print("Testing score: ",rf_clf.score(X_test, y_test))

print('Accuracy:', rf_clf.score(X, y) * 100)

# Confusion Matrix
conf_matrix = confusion_matrix(y_test, y_pred_rf)
print("\nConfusion Matrix:")
print(conf_matrix)

# Visualize Confusion Matrix
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix - Random Forest Classifier')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()

# Print results
print("Mean Squared Error:", mse)
print("R-squared Score:", r2)

# Graph
sap = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred_rf}).head(10)
sap.plot(kind='bar')
plt.grid(which='major', linewidth='0.8')
plt.grid(which='minor', linewidth='0.8')
plt.xticks(rotation=90)
plt.show()
```

**13.Kmeans-Iris**

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.cm as cm
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.metrics import confusion_matrix, classification_report, mean_squared_error, r2_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split

# Load Iris dataset
iris = load_iris()
x = iris.data

# Find the optimum number of clusters using the Elbow Method
sse = []  # Sum of Squared Error
for k in range(1, 11):
    km = KMeans(n_clusters=k, random_state=2)
    km.fit(x)
    sse.append(km.inertia_)

# Plot the Elbow Method
sns.set_style("whitegrid")
plt.figure(figsize=(8, 6))
g = sns.lineplot(x=range(1, 11), y=sse, marker='o')
g.set(xlabel="Number of clusters (k)", ylabel="Sum Squared Error", title='Elbow Method')
plt.xticks(range(1, 11))
plt.show()

# Perform KMeans clustering with the optimum number of clusters
optimal_k = 3
kmeans = KMeans(n_clusters=optimal_k, random_state=2)
kmeans.fit(x)

# Get cluster centers and predictions
cluster_centers = kmeans.cluster_centers_
y_pred = kmeans.fit_predict(x)

# Plot the clusters
plt.figure(figsize=(12, 5))

# Plot for petal features
plt.subplot(1, 2, 1)
plt.scatter(x[:, 2], x[:, 3], c=y_pred, cmap=cm.Accent)
plt.scatter(cluster_centers[:, 2], cluster_centers[:, 3], marker='^', c='red', s=200)
plt.title("Petal Features")
plt.xlabel("Petal Length (cm)")
plt.ylabel("Petal Width (cm)")
plt.grid(True)

# Plot for sepal features
plt.subplot(1, 2, 2)
plt.scatter(x[:, 0], x[:, 1], c=y_pred, cmap=cm.Accent)
plt.scatter(cluster_centers[:, 0], cluster_centers[:, 1], marker='^', c='red', s=200)
plt.title("Sepal Features")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Sepal Width (cm)")
plt.grid(True)

plt.tight_layout()
plt.show()

# Add confusion matrix and classification report
# Assume true labels are available (e.g., Iris species)
true_labels = iris.target

# Calculate confusion matrix and classification report
conf_matrix = confusion_matrix(true_labels, y_pred)
class_report = classification_report(true_labels, y_pred)

# Display confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=iris.target_names, yticklabels=iris.target_names)
plt.title('Confusion Matrix - KMeans Clustering')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()

# Display classification report
print("Classification Report:")
print(class_report)

# Actual and predicted
y_test = true_labels
y_pred_rf = y_pred  # Assuming y_pred_rf is used for KMeans predictions

# Evaluate the model (Note: These metrics are typically for regression tasks, not clustering)
mse = mean_squared_error(y_test, y_pred_rf)
r2 = r2_score(y_test, y_pred_rf)

# Print results
print("Mean Squared Error:", mse)
print("R-squared Score:", r2)

# Graph for the first 10 samples
plt.figure(figsize=(10, 6))
sap = pd.DataFrame({'Actual': y_test[:10], 'Predicted': y_pred_rf[:10]})
sap.plot(kind='bar')
plt.title('Actual vs Predicted - First 10 Samples')
plt.xlabel('Sample Index')
plt.ylabel('Class Label')
plt.show()
```

**16.DNN**

```python
import matplotlib.pyplot as plt
import tensorflow as tf
from keras.datasets import fashion_mnist
from keras.models import Sequential
from keras.layers import Flatten, Dense
import numpy as np
import seaborn as sns
from sklearn.metrics import classification_report

# Load Fashion MNIST dataset
(x_train, y_train), (x_test, y_test) = fashion_mnist.load_data()

# Display dataset information
print("Train shapes:", x_train.shape, y_train.shape)
print("Test shapes:", x_test.shape, y_test.shape)

# Display sample images
plt.imshow(x_train[0], cmap='Greys')
plt.figure(figsize=(18, 18))
class_labels = ["T-shirt/top", "Trouser", "Pullover", "Dress", "Coat", "Sandal", "Shirt", "Sneaker", "Bag", "Ankle boot"]

for i in range(25):
    plt.subplot(5, 5, i + 1)
    plt.imshow(x_train[i], cmap="gray")
    plt.axis('off')
    plt.title(class_labels[y_train[i]] + "=" + str(y_train[i]), fontsize=20)

# Normalize pixel values
x_train = x_train / 255.0
x_test = x_test / 255.0

# Define the DNN model
model = Sequential([
    Flatten(input_shape=(28, 28)),
    Dense(512, activation='relu'),
    Dense(256, activation='relu'),
    Dense(68, activation='relu'),
    Dense(10, activation='softmax')
])
model.summary()

# Compile and train the model
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.fit(x_train, y_train, epochs=10)

# Evaluate the model on the test data
test_loss, test_acc = model.evaluate(x_test, y_test)
print('Test accuracy:', test_acc)

# Make predictions
y_pred = model.predict(x_test)

# Display sample predictions
plt.figure(figsize=(16, 16))
for i in range(25):
    plt.subplot(5, 5, i + 1)
    plt.imshow(x_test[i], cmap="gray")
    plt.axis('off')
    plt.title("Actual= {} \n Predicted = {}".format(class_labels[y_test[i]], class_labels[np.argmax(y_pred[i])]))

# Display confusion matrix
cm = tf.math.confusion_matrix(y_test, [np.argmax(i) for i in y_pred])
sns.heatmap(cm, annot=True, fmt="d")

# Display classification report
cr = classification_report(y_test, [np.argmax(i) for i in y_pred], target_names=class_labels)
print(cr)
```

**17,18,19.NLP(Education is the most powerful weapon. That can be used to change the world)**

```python
import nltk
nltk.download('punkt')
import nltk
nltk.download('stopwords')
import nltk
nltk.download('wordnet')
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk import ngrams

# Given sentence
sentence = "Education is the most powerful weapon. That can be used to change the world"

# Convert the sentence to lowercase
lower_case = sentence.lower()
print(lower_case)

# Sentence Tokenization
s_tokens = sent_tokenize(lower_case)
print("Sentence Tokenization:\n", s_tokens)

# Word Tokenization
w_tokens = word_tokenize(lower_case)
print("Word Tokenization:\n", w_tokens)

# Remove Stopwords
stop_words = set(stopwords.words('english'))
filtered_text = [word for word in w_tokens if word not in stop_words]
print("Text after removing stop words:\n", filtered_text)

# Stemming
stemmer = PorterStemmer()
stemmed_words = [stemmer.stem(word) for word in w_tokens]
print("Sentence after stemming:\n", stemmed_words)

# Lemmatization
lemmatizer = WordNetLemmatizer()
lemmatized_words = [lemmatizer.lemmatize(word, pos='v') for word in w_tokens]
print("Sentence after lemmatization:\n", lemmatized_words)

# N-grams
words = word_tokenize(sentence)
uni_grams = list(ngrams(words, 1))
bi_grams = list(ngrams(words, 2))
tri_grams = list(ngrams(words, 3))

print("Uni-grams:")
for gram in uni_grams:
    print(gram)

print("\nBi-grams:")
for gram in bi_grams:
    print(gram)

print("\nTri-grams:")
for gram in tri_grams:
    print(gram)
```
